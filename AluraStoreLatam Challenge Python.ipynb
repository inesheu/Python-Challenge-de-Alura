### Importación de datos



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt
```

```python
# Sacar ingresos de tiendas sumando la columna 'Precio'
ingresos1 = sum(tienda1['Precio'])
ingresos2 = sum(tienda2['Precio'])
ingresos3 = sum(tienda3['Precio'])
ingresos4 = sum(tienda4['Precio'])

#Guardar ingresos en un diccionario
ingresosPorTienda = {
    'Tienda 1': ingresos1,
    'Tienda 2': ingresos2,
    'Tienda 3': ingresos3,
    'Tienda 4': ingresos4
}

#Mostrar resultados
print(f'Ingresos por tienda: {ingresosPorTienda}')
```

```python
#Dibujar grafico de barras
plt.bar(ingresosPorTienda.keys(), ingresosPorTienda.values())
plt.title('Ventas por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ventas')
plt.show()
```

# 2. Ventas por categoría

```python
# TIENDA 1
# Crear diccionario vacio de ventas por categoria
ventasPorCategoria1 = {}

# Crear lista de todas las categorias
categorias1 = list(tienda1['Categoría del Producto'].unique())

# Recorrer cada categoria
for categoria in categorias:
  suma = 0
  for i in range(len(tienda1)):
    if tienda1['Categoría del Producto'][i] == categoria:
      suma += tienda1['Precio'][i]
  ventasPorCategoria1[categoria] = float(suma)

# Mostrar resultados Tienda 1
print(f'Ventas por categoria Tienda 1: {ventasPorCategoria1}')
```

```python
# Crear grafico por categoria
plt.bar(ventasPorCategoria1.keys(), ventasPorCategoria1.values())

# Arreglar grafico
plt.xticks(rotation=45, ha='right')
plt.title('Ventas por Categoria Tienda 1')
plt.xlabel('Categoria')
plt.ylabel('Ventas')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacionTienda1 = tienda1['Calificación'].mean()
calificacionTienda2 = tienda2['Calificación'].mean()
calificacionTienda3 = tienda3['Calificación'].mean()
calificacionTienda4 = tienda4['Calificación'].mean()



calificaciones = {
    'Tienda 1': float(calificacionTienda1),
    'Tienda 2': float(calificacionTienda2),
    'Tienda 3': float(calificacionTienda3),
    'Tienda 4': float(calificacionTienda4)
}


print(f'Calificacion por tienda: {calificaciones}')
```

```python
plt.bar(calificaciones.keys(), calificaciones.values())
plt.title('Calificacion promedio por tienda')
plt.ylabel('Calificacion promedio')
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Tienda 1 - Calcular top 5 y bottom 5 productos vendidos
top5_tienda1 = tienda1.groupby('Producto')['Precio'].sum().sort_values(ascending=False).head(5)
bottom5_tienda1 = tienda1.groupby('Producto')['Precio'].sum().sort_values(ascending=True).head(5)# Graficar Top 5 en verde
plt.bar(top5_tienda1.index, top5_tienda1.values, color='green', label='Top 5 Productos')

# Graficar Bottom 5 en rojo
plt.bar(bottom5_tienda1.index, bottom5_tienda1.values, color='red', label='Bottom 5 Productos')

# Añadir etiquetas y título
plt.xlabel('Productos')
plt.ylabel('Ventas Totales (Precio)')
plt.title('Productos Mas y Menos Vendidos - Tienda 1')

# Rotar etiquetas de los productos para que sean legibles
plt.xticks(rotation=45, ha='right')

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()

# Resultados
print(f'Top 5 productos vendidos Tienda 1: {top5_tienda1}')
print(f'Bottom 5 productos vendidos Tienda 1: {bottom5_tienda1}')
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio por tienda
costoEnvioTienda1 = tienda1['Costo de envío'].mean()
costoEnvioTienda2 = tienda2['Costo de envío'].mean()
costoEnvioTienda3 = tienda3['Costo de envío'].mean()
costoEnvioTienda4 = tienda4['Costo de envío'].mean()

# Crear un diccionario con los resultados
costoEnvioPromedio = {
    'Tienda 1': costoEnvioTienda1,
    'Tienda 2': costoEnvioTienda2,
    'Tienda 3': costoEnvioTienda3,
    'Tienda 4': costoEnvioTienda4
}

# Datos de costo de envío promedio (ya calculado)
tiendas = list(costoEnvioPromedio.keys())
costos_envio = list(costoEnvioPromedio.values())

# Crear el gráfico de barras
plt.bar(tiendas, costos_envio)

# Añadir título y etiquetas
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio')

# Mostrar el gráfico
plt.show()

# Resultados
print(f'Top 5 productos vendidos Tienda 4: {top5_tienda4}')
print(f'Bottom 5 productos vendidos Tienda 4: {bottom5_tienda4}')
```
